```python
import pandas as pd
```

```python
df = pd.read_csv('data\\Kangaroo.csv')
df.head()
```

```python
df.info()
```

```python
df_subset=df[["subtype","bedroomCount","bathroomCount","postCode","habitableSurface","gardenSurface","hasAttic","hasTerrace","hasBasement","buildingCondition","epcScore","facedeCount","price"]]

df_subset.head()
```

```python
import numpy as np

belgian_postcode_details=pd.read_csv('georef-belgium-postal-codes.csv',sep=';')
belgian_postcode_geolocation=belgian_postcode_details[["Geo Point","Post code"]]
belgian_postcode_geolocation.head()

dict_belgian_postcodes_longitude={}
dict_belgian_postcodes_latitude={}
for index, row in belgian_postcode_geolocation.iterrows():
    if type(row["Geo Point"])!=float:#weird stuff handling
        lat,lon=np.array(row["Geo Point"].split(","),dtype=float)#Extract latitude and longitude
        dict_belgian_postcodes_latitude[row["Post code"]]=lat#place longitude
        dict_belgian_postcodes_longitude[row["Post code"]]=lon#place latitude

print(dict_belgian_postcodes_latitude)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn import neighbors

#We drop place with no info about price and habitableSurface
df_subset.dropna(axis=0,subset=['price','habitableSurface','bedroomCount','bathroomCount','epcScore','facedeCount','buildingCondition'],inplace=True)

#Getting the price/sqm for each subset
df_subset["longitude"]=df_subset['postCode'].map(dict_belgian_postcodes_longitude)
df_subset["latitude"]=df_subset['postCode'].map(dict_belgian_postcodes_latitude)

dict_state={'AS_NEW':0,'JUST_RENOVATED':1,'GOOD':2,'TO_RENOVATE':3,'TO_RESTORE':4 ,'TO_BE_DONE_UP':5}
dict_epc={'A++':0,'A+':1, 'A':2,'B':3,'C':4, 'D':5, 'E':6, 'F':7,'G':8}

df_subset[["gardenSurface","hasAttic","hasTerrace","hasBasement"]]=df_subset[["gardenSurface","hasAttic","hasTerrace","hasBasement"]].fillna(0).astype(int)
df_subset['epcScore_number']=df_subset['epcScore'].map(dict_epc)
df_subset['buildingCondition_number']=df_subset['buildingCondition'].map(dict_state)
df_subset.drop(['epcScore','buildingCondition'], axis=1, inplace=True)
df_subset.dropna(inplace=True)
df_subset=pd.get_dummies(df_subset)
df_subset=df_subset[df_subset['price']<1500000]

#We save a subset of the data to get stats about geographical costs belgium
map_subset,df_subset=train_test_split(df_subset,shuffle=True,random_state=42,train_size=0.1)

map_subset["price_per_sqm"]=map_subset["price"]/(map_subset['habitableSurface']+0.3*map_subset["gardenSurface"])

df_subset["price_per_sqm"]=neighbors.KNeighborsRegressor(20, weights="uniform").fit(map_subset[["longitude","latitude"]], map_subset['price_per_sqm']).predict(df_subset[["longitude","latitude"]])

df_subset["expected_price"]=df_subset["price_per_sqm"]*(df_subset["habitableSurface"]+0.3*df_subset["gardenSurface"])
```

```python
#We split whats left into df_subset and target
X=df_subset.drop(["price"],inplace=False, axis=1)
y=df_subset["price"]


#Do the final split for train_test etc
X_train,X_test,y_train,y_test=train_test_split(X,y,shuffle=True,random_state=42,train_size=0.8)
```

```python
X_train.head()
```

```python
print(np.mean(np.abs(X_test["expected_price"]-y_test)))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
pred=reg.predict(X_test)
print(np.mean(np.abs(pred-y_test)))
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
gboost=HistGradientBoostingRegressor(learning_rate=0.5,
    max_depth=4, random_state=0).fit(X_train, y_train)

pred=gboost.predict(X_test)
print(np.mean(np.abs(pred-y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=5, random_state=0)
clf.fit(X_train, y_train)

pred=clf.predict(X_test)
print(np.mean(np.abs(pred-y_test)))
```
